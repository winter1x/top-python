#### **Метод describe()**

В библиотеке Pandas для описания **основных характеристик** существует метод `describe()`.

**Для числовых данных describe выводит следующие характеристики:**
1. `count` - общее количество элементов

2. `mean` - среднее значение

3. `std` - среднеквадратическое отклонение
4. `min` - минимальное значение
5. `25%` - первый квартиль (Q1)
6. `50%` - второй квартиль, медиана (Q2)
7. `75%` - третий квартиль (Q3)
8. `max` - максимальное значение
